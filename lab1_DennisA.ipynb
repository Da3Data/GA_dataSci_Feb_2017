Extract the subset of rows of the data frame where Ozone values are above 31 and Temp values
are above 90. What is the mean of Solar.R in this subset?

```python
## broken down into 2 steps

# Step-1
data.loc[(data['Ozone'] > 31) & (data['Temp'] > 90), ['Ozone','Temp','Solar.R']]
```

```python
# Step-2
# What is the mean of Solar.R in this subset?
# dataframe['column_name'].mean()

SolarR_mean = (data.loc[(data['Ozone'] > 31) & (data['Temp'] > 90), ['Ozone','Temp','Solar.R']])

SolarR_mean['Solar.R'].mean()
```

What is the mean of "Temp" when "Month" is equal to 6?

```python
# Step-1
# month = 6
data.loc[(data['Month'] == 6) , ['Month','Temp']]
```

```python
# Step-2
# month = 6
Mean_Temp_June = (data.loc[(data['Month'] == 6) , ['Month','Temp']])

Mean_Temp_June['Temp'].mean()
```

What was the maximum ozone value in the month of May (i.e. Month = 5)?
